```python
import zipfile
import os

zip_path = "/content/gtzan-dataset-music-genre-classification.zip"

extract_path = "/content/data"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Extraction complete")
```

```python
import os
import numpy as np
import librosa
from tqdm import tqdm

DATA_PATH = "/content/data/Data/genres_original"
```

```python
audio_data = []
genres_list = []

genres = os.listdir(DATA_PATH)

for genre in genres:
    genre_folder = os.path.join(DATA_PATH, genre)

    for file in tqdm(os.listdir(genre_folder)):
        if file.endswith(".wav"):
            file_path = os.path.join(genre_folder, file)

            try:
                audio, sr = librosa.load(file_path, sr=22050)

                if len(audio) > 22050 * 30:
                    audio = audio[:22050 * 30]
                else:
                    audio = np.pad(audio, (0, 22050 * 30 - len(audio)))

                audio = (audio - np.mean(audio)) / (np.std(audio) + 1e-6)

                audio_data.append(audio)
                genres_list.append(genre)

            except Exception as e:
                print("ERROR:", file_path)

print("Total samples:", len(audio_data))
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(genres_list)

print("Classes:", le.classes_)
```

```python
from sklearn.model_selection import train_test_split

indices = np.arange(len(audio_data))

train_idx, test_idx = train_test_split(
    indices,
    test_size=0.2,
    stratify=genres_list,
    random_state=42
)

def create_mel_segments(indices):
    mel_segments = []
    mel_labels = []

    segment_length = 22050 * 5

    for i in indices:
        signal = audio_data[i]
        genre = genres_list[i]
        hop_length = segment_length // 2

        for start in range(0, len(signal), segment_length):
            segment = signal[start:start + segment_length]

            if len(segment) < segment_length:
                continue

            mel = librosa.feature.melspectrogram(
                y=segment,
                sr=22050,
                n_mels=96
            )
            mel_db = librosa.power_to_db(mel)

            mfcc = librosa.feature.mfcc(y=segment, sr=22050, n_mfcc=20)
            chroma = librosa.feature.chroma_stft(y=segment, sr=22050)

            mfcc = librosa.util.fix_length(mfcc, size=mel_db.shape[1])
            chroma = librosa.util.fix_length(chroma, size=mel_db.shape[1])

            combined = np.vstack([mel_db, mfcc, chroma])

            mel_segments.append(combined)
            mel_labels.append(genre)

    return np.array(mel_segments), np.array(mel_labels)


train_mels, train_labels = create_mel_segments(train_idx)
test_mels, test_labels = create_mel_segments(test_idx)

print("Train shape:", train_mels.shape)
print("Test shape:", test_mels.shape)
```

```python
import torch
from torch.utils.data import Dataset

class AttentionDataset(Dataset):
    def __init__(self, mels, labels, le):
        self.mels = mels
        self.labels = le.transform(labels)

    def __len__(self):
        return len(self.mels)

    def __getitem__(self, idx):
      mel = self.mels[idx]
      mel = (mel - np.min(mel)) / (np.max(mel) - np.min(mel) + 1e-6)

      mel = torch.tensor(mel, dtype=torch.float32)
      if mel.ndim == 2:
          mel = mel.unsqueeze(0)

      label = torch.tensor(self.labels[idx], dtype=torch.long)

      return mel, label
```

```python
from torch.utils.data import DataLoader

train_ds = AttentionDataset(train_mels, train_labels, le)
test_ds  = AttentionDataset(test_mels, test_labels, le)

train_dl = DataLoader(train_ds, batch_size=32, shuffle=True)
test_dl  = DataLoader(test_ds, batch_size=32)
```

```python
import torch.nn as nn
class AttentionModel(nn.Module):
    def __init__(self, num_classes=10):
        super().__init__()

        self.cnn = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout(0.2),

            nn.Conv2d(32, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout(0.2),

            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout(0.3),

            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout(0.3)
        )
        self.gru = nn.GRU(
            input_size=256,
            hidden_size=256,
            batch_first=True,
            bidirectional=True
        )

        self.attn = nn.Sequential(
            nn.Linear(512, 128),
            nn.Tanh(),
            nn.Linear(128, 1)
        )

        self.dropout = nn.Dropout(0.3)
        self.fc = nn.Linear(512, num_classes)

    def forward(self, x):
        x = self.cnn(x)
        B, C, H, W = x.shape
        x = x.view(B, C, H * W)
        x = x.permute(0, 2, 1)
        gru_out, _ = self.gru(x)

        attn_weights = torch.softmax(self.attn(gru_out), dim=1)
        context = torch.sum(attn_weights * gru_out, dim=1)

        context = self.dropout(context)
        return self.fc(context)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = AttentionModel(num_classes=10).to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(),lr=0.0005, weight_decay=1e-4)

scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode='min',
    patience=3,
    factor=0.5
)

EPOCHS = 100

for epoch in range(EPOCHS):
    model.train()
    total_loss = 0

    for x, y in train_dl:
        x, y = x.to(device), y.to(device)

        out = model(x)
        loss = criterion(out, y)

        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()

        total_loss += loss.item()
    scheduler.step(total_loss)

    print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")
```

```python
from sklearn.metrics import accuracy_score, f1_score

model.eval()
preds = []
true = []

with torch.no_grad():
    for x, y in test_dl:
        x = x.to(device)

        out = model(x)
        pred = torch.argmax(out, dim=1)

        preds.extend(pred.cpu().numpy())
        true.extend(y.numpy())

print("Accuracy:", accuracy_score(true, preds))
print("F1 Score:", f1_score(true, preds, average='macro'))
```
