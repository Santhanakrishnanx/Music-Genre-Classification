```python
import kagglehub

path = kagglehub.dataset_download("andradaolteanu/gtzan-dataset-music-genre-classification")
print("Path:", path)
```

```python
import os
import librosa

X = []
y = []

audio_path = f"{path}/Data/genres_original"
genres = os.listdir(audio_path)

for genre in genres:
    genre_folder = os.path.join(audio_path, genre)

    for file in os.listdir(genre_folder):
        if file.endswith(".wav"):
            file_path = os.path.join(genre_folder, file)

            try:
                audio, sr = librosa.load(file_path, sr=24000)
                X.append(audio)
                y.append(genre)

            except Exception as e:
                print(f"Skipping bad file: {file_path}")
```

```python
import torch.nn as nn

class FullModel(nn.Module):
    def __init__(self, mert_model, num_classes=10):
        super().__init__()
        self.mert = mert_model

        self.gru = nn.GRU(
            input_size=1536,
            hidden_size=256,
            batch_first=True,
            bidirectional=True
        )

        self.attn = nn.Sequential(
            nn.Linear(512, 128),
            nn.Tanh(),
            nn.Linear(128, 1)
        )

        self.dropout = nn.Dropout(0.3)
        self.fc = nn.Linear(512, num_classes)

    def forward(self, x):
        B = x.shape[0]

        inputs = processor(
            x.detach().cpu().numpy(),
            sampling_rate=24000,
            return_tensors="pt",
            padding=True
        )

        inputs = {k: v.to(x.device) for k, v in inputs.items()}

        outputs = self.mert(**inputs)
        hidden = outputs.last_hidden_state  # (B, T, D)

        mean_pool = hidden.mean(dim=1)
        max_pool = hidden.max(dim=1).values

        emb = torch.cat([mean_pool, max_pool], dim=1)  # (B, 1536)
        emb = emb.unsqueeze(1)  # (B, 1, 1536)
        gru_out, _ = self.gru(emb)

        attn_weights = torch.softmax(self.attn(gru_out), dim=1)
        attn_context = torch.sum(attn_weights * gru_out, dim=1)
        max_context = torch.max(gru_out, dim=1).values

        context = attn_context + max_context

        context = self.dropout(context)
        return self.fc(context)
```

```python
from transformers import AutoProcessor, AutoModel
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

processor = AutoProcessor.from_pretrained("m-a-p/MERT-v1-95M")
mert_model = AutoModel.from_pretrained("m-a-p/MERT-v1-95M").to(device)
model = FullModel(mert_model).to(device)

model.train()
```

```python
def preprocess_audio(audio, max_length=24000*5):
    if len(audio) > max_length:
        audio = audio[:max_length]
    return audio


def extract_embedding(audio):
    audio = preprocess_audio(audio)

    inputs = processor(
        audio,
        sampling_rate=24000,
        return_tensors="pt",
        padding=True
    )

    inputs = {k: v.to(device) for k, v in inputs.items()}
    outputs = mert_model(**inputs)
    hidden = outputs.last_hidden_state
    mean_pool = hidden.mean(dim=1)
    max_pool = hidden.max(dim=1).values

    embedding = torch.cat([mean_pool, max_pool], dim=1)
    return embedding.detach().squeeze().cpu().numpy()
```

```python
import numpy as np

X_seq = []
y_seq = []

segment_len = 24000 * 5  # 5 sec

for i in range(len(X)):
    audio = X[i]
    segments = []

    for start in range(0, len(audio) - segment_len, segment_len):
        segment = audio[start:start + segment_len]
        emb = extract_embedding(segment)
        segments.append(emb)

    if len(segments) > 0:
        X_seq.append(np.array(segments))
        y_seq.append(y[i])

    if i % 50 == 0:
        print(f"Processed {i}/{len(X)}")
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# split FIRST
X_train_raw, X_test_raw, y_train_raw, y_test_raw = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

# THEN fit encoder on full labels
le = LabelEncoder()
le.fit(y)
```

```python
import torch
from torch.utils.data import Dataset

class RawAudioDataset(Dataset):
    def __init__(self, X, y, le):
        self.X = X
        self.y = le.transform(y)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        audio = self.X[idx]
        max_len = 24000 * 5
        if len(audio) > max_len:
            audio = audio[:max_len]
        else:
            audio = np.pad(audio, (0, max_len - len(audio)))

        audio = torch.tensor(audio, dtype=torch.float32)

        return audio, torch.tensor(self.y[idx])
```

```python
from torch.utils.data import DataLoader

train_ds = RawAudioDataset(X_train_raw, y_train_raw, le)
test_ds  = RawAudioDataset(X_test_raw, y_test_raw, le)

train_dl = DataLoader(train_ds, batch_size=8, shuffle=True)
test_dl  = DataLoader(test_ds, batch_size=8)
```

```python
import torch.nn as nn

class MERT_Attention_Model(nn.Module):
    def __init__(self, input_size=1536, num_classes=10):
        super().__init__()

        self.gru = nn.GRU(
            input_size=input_size,
            hidden_size=256,
            batch_first=True,
            bidirectional=True
        )

        self.attn = nn.Sequential(
            nn.Linear(512, 128),
            nn.Tanh(),
            nn.Linear(128, 1)
        )

        self.dropout = nn.Dropout(0.3)
        self.fc = nn.Linear(512, num_classes)

    def forward(self, x):
        gru_out, _ = self.gru(x)
        attn_weights = torch.softmax(self.attn(gru_out), dim=1)
        context = torch.sum(attn_weights * gru_out, dim=1)

        context = self.dropout(context)
        return self.fc(context)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = FullModel(mert_model).to(device)
for name, param in model.mert.named_parameters():
    if "encoder.layers.0" in name or "encoder.layers.1" in name:
        param.requires_grad = False

criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam([
    {'params': model.mert.parameters(), 'lr': 5e-6},
    {'params': list(model.gru.parameters()) +
               list(model.attn.parameters()) +
               list(model.fc.parameters()),
     'lr': 3e-4}
], weight_decay=1e-4)

scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode='min',
    patience=3,
    factor=0.5
)

EPOCHS = 100

for epoch in range(EPOCHS):
    model.train()
    total_loss = 0

    for x, y in train_dl:
        x, y = x.to(device), y.to(device)

        with torch.cuda.amp.autocast():
            out = model(x)
            loss = criterion(out, y)

        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()

        total_loss += loss.item()
    scheduler.step(total_loss)

    print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")
```

```python
from sklearn.metrics import accuracy_score, f1_score

model.eval()
preds = []
true = []

with torch.no_grad():
    for x, y in test_dl:
        x = x.to(device)

        out = model(x)
        pred = torch.argmax(out, dim=1)

        preds.extend(pred.cpu().numpy())
        true.extend(y.numpy())

print("Accuracy:", accuracy_score(true, preds))
print("F1 Score:", f1_score(true, preds, average='macro'))
```
