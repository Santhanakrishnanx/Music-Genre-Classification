```python
import kagglehub

path = kagglehub.dataset_download("andradaolteanu/gtzan-dataset-music-genre-classification")
print("Path:", path)
```

```python
import os
import librosa

X = []
y = []

audio_path = f"{path}/Data/genres_original"
genres = os.listdir(audio_path)

for genre in genres:
    genre_folder = os.path.join(audio_path, genre)

    for file in os.listdir(genre_folder):
        if file.endswith(".wav"):
            file_path = os.path.join(genre_folder, file)

            try:
                audio, sr = librosa.load(file_path, sr=24000)
                X.append(audio)
                y.append(genre)

            except Exception as e:
                print(f"Skipping bad file: {file_path}")
```

```python
from transformers import AutoProcessor, AutoModel
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

processor = AutoProcessor.from_pretrained("m-a-p/MERT-v1-95M")
model = AutoModel.from_pretrained("m-a-p/MERT-v1-95M").to(device)

model.eval()
```

```python
def preprocess_audio(audio, max_length=16000*5):
    if len(audio) > max_length:
        audio = audio[:max_length]
    return audio


def extract_embedding(audio):
    audio = preprocess_audio(audio)

    inputs = processor(
        audio,
        sampling_rate=24000,
        return_tensors="pt",
        padding=True
    )

    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        outputs = model(**inputs)

    embedding = outputs.last_hidden_state.mean(dim=1) + \
            outputs.last_hidden_state.max(dim=1).values
    return embedding.squeeze().cpu().numpy()
```

```python
print(sr)
X_emb = []
y_emb = []

for i in range(len(X)):
    emb = extract_embedding(X[i])

    X_emb.append(emb)
    y_emb.append(y[i])

    if i % 50 == 0:
        print(f"Processed {i}/{len(X)}")
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

le = LabelEncoder()
y_encoded = le.fit_transform(y_emb)
X_train, X_test, y_train, y_test = train_test_split(
    X_emb, y_encoded, test_size=0.2, random_state=42
)
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(
    hidden_layer_sizes=(512, 256, 128),
    activation='relu',
    solver='adam',
    alpha=1e-4,
    batch_size=32,
    learning_rate_init=0.001,
    max_iter=80,
    random_state=42
)

mlp.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

y_pred = mlp.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred, average='macro'))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("MERT + MLP Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

# GAN Based Model


```python
import os
import numpy as np
import librosa

labels = []
audio_data = []
genres_list = []

audio_path = f"{path}/Data/genres_original"
genres = os.listdir(audio_path)

for genre in genres:
    genre_folder = os.path.join(audio_path, genre)

    for file in os.listdir(genre_folder):
        if file.endswith(".wav"):
            file_path = os.path.join(genre_folder, file)

            try:
                audio, sr = librosa.load(file_path, sr=22050)
                if len(audio) > 22050 * 30:
                    audio = audio[:22050 * 30]
                else:
                    audio = np.pad(audio, (0, 22050 * 30 - len(audio)))
                audio = (audio - np.mean(audio)) / (np.std(audio) + 1e-6)
                audio_data.append(audio)
                genres_list.append(genre)


                mel = librosa.feature.melspectrogram(
                    y=audio, sr=22050, n_mels=128
                )
                mel_db = librosa.power_to_db(mel)

                save_file = f"{genre}_{file}.npy"
                np.save(save_file, mel_db)

                labels.append((save_file, genre, file_path))

            except Exception as e:
                print("ERROR:", file_path, e)

print("Total samples:", len(audio_data))
```

```python
import numpy as np
import pywt
import librosa
import numpy as np
import pywt
import librosa

def extract_features_segmented(audio_data, genres_list):
    X = []
    y = []

    segment_length = 22050 * 5  # 5 sec segments

    for i in range(len(audio_data)):
        signal = audio_data[i]
        genre = genres_list[i]

        # split into 5-second chunks
        for start in range(0, len(signal), segment_length):
            segment = signal[start:start + segment_length]

            if len(segment) < segment_length:
                continue

            # =========================
            # MFCC
            # =========================
            mfcc = librosa.feature.mfcc(y=segment, sr=22050, n_mfcc=20)
            mfcc_mean = np.mean(mfcc, axis=1)
            mfcc_std = np.std(mfcc, axis=1)

            # =========================
            # FFT
            # =========================
            fft = np.abs(np.fft.fft(segment))[:3000]
            fft_mean = np.mean(fft)
            fft_std = np.std(fft)

            # =========================
            # Spectral
            # =========================
            spec_centroid = np.mean(librosa.feature.spectral_centroid(y=segment, sr=22050))
            spec_bandwidth = np.mean(librosa.feature.spectral_bandwidth(y=segment, sr=22050))
            spec_rolloff = np.mean(librosa.feature.spectral_rolloff(y=segment, sr=22050))
            zero_cross = np.mean(librosa.feature.zero_crossing_rate(segment))

            # =========================
            # Wavelet
            # =========================
            coeffs = pywt.wavedec(segment, 'db4', level=4)

            wavelet_features = []
            for c in coeffs:
                wavelet_features.extend([
                    np.mean(c),
                    np.std(c)
                ])

            # =========================
            # Final feature
            # =========================
            features = np.concatenate([
                mfcc_mean,
                mfcc_std,
                [fft_mean, fft_std],
                [spec_centroid, spec_bandwidth, spec_rolloff, zero_cross],
                wavelet_features
            ])

            X.append(features)
            y.append(genre)

    return np.array(X), np.array(y)

X_feat, y_feat = extract_features_segmented(audio_data, genres_list)

print("FINAL Feature shape:", X_feat.shape)
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

le = LabelEncoder()
y_encoded = le.fit_transform(y_feat)

print("Classes:", le.classes_)
X_train, X_test, y_train, y_test = train_test_split(
    X_feat,
    y_encoded,
    test_size=0.2,
    stratify=y_encoded,
    random_state=42
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train = np.clip(X_train, -3, 3)
X_test = np.clip(X_test, -3, 3)

print("Final Train shape:", X_train.shape)
```

```python
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# =========================
# DATA
# =========================
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.long)

X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.long)

train_ds = TensorDataset(X_train_tensor, y_train_tensor)
test_ds = TensorDataset(X_test_tensor, y_test_tensor)

train_dl = DataLoader(train_ds, batch_size=32, shuffle=True)
test_dl = DataLoader(test_ds, batch_size=32)

# =========================
# CLASS WEIGHTS (IMPORTANT)
# =========================
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)

class_weights = torch.tensor(class_weights, dtype=torch.float32).to(device)

# =========================
# STRONGER MODEL
# =========================
class MLP(nn.Module):
    def __init__(self, input_dim, num_classes):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Dropout(0.4),

            nn.Linear(512, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.4),

            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),

            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        return self.net(x)

model = MLP(X_train.shape[1], 10).to(device)

criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = torch.optim.Adam(model.parameters(), lr=0.0005)

# =========================
# TRAINING (WITH EARLY STOP)
# =========================
EPOCHS = 100

best_loss = float('inf')
patience = 10
counter = 0

for epoch in range(EPOCHS):
    model.train()
    total_loss = 0

    for x, y in train_dl:
        x, y = x.to(device), y.to(device)

        out = model(x)
        loss = criterion(out, y)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")

    # early stopping
    if total_loss < best_loss:
        best_loss = total_loss
        counter = 0
    else:
        counter += 1

    if counter >= patience:
        print("Early stopping triggered")
        break
```

```python
from sklearn.metrics import accuracy_score, f1_score

model.eval()
preds = []
true = []

with torch.no_grad():
    for x, y in test_dl:
        x = x.to(device)

        out = model(x)
        pred = torch.argmax(out, dim=1)

        preds.extend(pred.cpu().numpy())
        true.extend(y.numpy())

print("Accuracy:", accuracy_score(true, preds))
print("F1 Score:", f1_score(true, preds, average='macro'))
```

```python
import numpy as np
import librosa

mel_segments = []
mel_labels = []

segment_length = 22050 * 5  # same as before

for i in range(len(audio_data)):
    signal = audio_data[i]
    genre = genres_list[i]

    for start in range(0, len(signal), segment_length):
        segment = signal[start:start + segment_length]

        if len(segment) < segment_length:
            continue

        mel = librosa.feature.melspectrogram(
            y=segment,
            sr=22050,
            n_mels=128
        )

        mel_db = librosa.power_to_db(mel)

        mel_segments.append(mel_db)
        mel_labels.append(genre)

mel_segments = np.array(mel_segments)

print("Mel segments shape:", mel_segments.shape)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn.functional as F

class MelDataset(Dataset):
    def __init__(self, mels, labels, le):
        self.mels = mels
        self.labels = le.transform(labels)

    def __len__(self):
        return len(self.mels)

    def __getitem__(self, idx):
        mel = self.mels[idx]

        mel = (mel - np.mean(mel)) / (np.std(mel) + 1e-6)

        mel = torch.tensor(mel, dtype=torch.float32).unsqueeze(0)

        mel = F.interpolate(
            mel.unsqueeze(0),
            size=(128, 128),
            mode='bilinear'
        ).squeeze(0)

        label = torch.tensor(self.labels[idx], dtype=torch.long)

        return mel, label


gan_dataset = MelDataset(mel_segments, mel_labels, le)
gan_loader = DataLoader(gan_dataset, batch_size=32, shuffle=True)
```

```python
import torch.nn as nn

class Generator(nn.Module):
    def __init__(self, z_dim=100):
        super().__init__()

        self.net = nn.Sequential(

            nn.Linear(z_dim, 256*8*8),
            nn.ReLU(),
            nn.Unflatten(1, (256, 8, 8)),

            nn.ConvTranspose2d(256, 128, 4, 2, 1),
            nn.ReLU(),

            nn.ConvTranspose2d(128, 64, 4, 2, 1),
            nn.ReLU(),

            nn.ConvTranspose2d(64, 1, 4, 2, 1),
            nn.Tanh()
        )

    def forward(self, z):
        return self.net(z)
```

```python
import torchvision.models as models

class Discriminator(nn.Module):
    def __init__(self, num_classes=10):
        super().__init__()

        self.backbone = models.resnet18(pretrained=True)
        self.backbone.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)

        self.features = nn.Sequential(*list(self.backbone.children())[:-1])

        self.adv_head = nn.Sequential(
            nn.Flatten(),
            nn.Linear(512, 1),
            nn.Sigmoid()
        )

        self.genre_head = nn.Linear(512, num_classes)

    def forward(self, x):
        feat = self.features(x)
        feat = feat.view(feat.size(0), -1)

        adv = self.adv_head(feat)
        genre = self.genre_head(feat)

        return adv, feat, genre
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

z_dim = 100

G = Generator(z_dim).to(device)
D = Discriminator(num_classes=10).to(device)

criterion_adv = nn.BCELoss()
criterion_cls = nn.CrossEntropyLoss()

g_optimizer = torch.optim.Adam(G.parameters(), lr=0.0002, betas=(0.5, 0.999))
d_optimizer = torch.optim.Adam(D.parameters(), lr=0.0002, betas=(0.5, 0.999))
```

```python
EPOCHS = 10

for epoch in range(EPOCHS):
    for real_imgs, labels in gan_loader:

        real_imgs = real_imgs.to(device)
        labels = labels.to(device)

        batch_size = real_imgs.size(0)

        real_label = torch.ones(batch_size, 1).to(device)
        fake_label = torch.zeros(batch_size, 1).to(device)

        # ===== TRAIN D =====
        z = torch.randn(batch_size, z_dim).to(device)
        fake_imgs = G(z)

        d_real, feat_real, genre_real = D(real_imgs)
        d_fake, _, _ = D(fake_imgs.detach())

        adv_loss = criterion_adv(d_real, real_label) + criterion_adv(d_fake, fake_label)
        cls_loss = criterion_cls(genre_real, labels)

        d_loss = adv_loss + cls_loss

        d_optimizer.zero_grad()
        d_loss.backward()
        d_optimizer.step()

        # ===== TRAIN G =====
        z = torch.randn(batch_size, z_dim).to(device)
        fake_imgs = G(z)

        d_fake, _, _ = D(fake_imgs)
        g_loss = criterion_adv(d_fake, real_label)

        g_optimizer.zero_grad()
        g_loss.backward()
        g_optimizer.step()

    print(f"Epoch {epoch+1} | D Loss: {d_loss.item():.4f} | G Loss: {g_loss.item():.4f}")
```

```python
def extract_gan_features(dataset, D, device):
    loader = DataLoader(dataset, batch_size=32)

    D.eval()
    feats = []

    with torch.no_grad():
        for x, _ in loader:
            x = x.to(device)

            _, f, _ = D(x)
            f = torch.nn.functional.normalize(f, dim=1)

            feats.append(f.cpu().numpy())

    return np.vstack(feats)


X_gan = extract_gan_features(gan_dataset, D, device)

print("GAN feature shape:", X_gan.shape)
```

```python
X_final = np.hstack([X_feat, X_gan])
print("Final feature shape:", X_final.shape)
```

```python
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# =========================
# DATA
# =========================
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.long)

X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.long)

train_ds = TensorDataset(X_train_tensor, y_train_tensor)
test_ds = TensorDataset(X_test_tensor, y_test_tensor)

train_dl = DataLoader(train_ds, batch_size=32, shuffle=True)
test_dl = DataLoader(test_ds, batch_size=32)

# =========================
# CLASS WEIGHTS (IMPORTANT)
# =========================
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)

class_weights = torch.tensor(class_weights, dtype=torch.float32).to(device)

# =========================
# STRONGER MODEL
# =========================
class MLP(nn.Module):
    def __init__(self, input_dim, num_classes):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Dropout(0.4),

            nn.Linear(512, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.4),

            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),

            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        return self.net(x)

model = MLP(X_train.shape[1], 10).to(device)

criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = torch.optim.Adam(model.parameters(), lr=0.0005)

# =========================
# TRAINING (WITH EARLY STOP)
# =========================
EPOCHS = 100

best_loss = float('inf')
patience = 10
counter = 0

for epoch in range(EPOCHS):
    model.train()
    total_loss = 0

    for x, y in train_dl:
        x, y = x.to(device), y.to(device)

        out = model(x)
        loss = criterion(out, y)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")

    # early stopping
    if total_loss < best_loss:
        best_loss = total_loss
        counter = 0
    else:
        counter += 1

    if counter >= patience:
        print("Early stopping triggered")
        break
```

```python
from sklearn.metrics import accuracy_score, f1_score

model.eval()
preds = []
true = []

with torch.no_grad():
    for x, y in test_dl:
        x = x.to(device)

        out = model(x)
        pred = torch.argmax(out, dim=1)

        preds.extend(pred.cpu().numpy())
        true.extend(y.numpy())

print("Accuracy:", accuracy_score(true, preds))
print("F1 Score:", f1_score(true, preds, average='macro'))
```
